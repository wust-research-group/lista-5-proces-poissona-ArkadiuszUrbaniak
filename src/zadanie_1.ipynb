```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Metoda 1


```python
def poiss1(lambd=2, T=10):
    S = []
    I = 0
    t = 0
    while t <= T:
        U = np.random.uniform(0, 1)
        t = t - 1 / lambd * np.log(U)
        if t <= T:
            I = I + 1
            S.append(t)
    n = len(S)
    N = np.arange(1, n + 1)
    return S, N


S, N = poiss1()

plt.step(S, N)
plt.title("Proces Poissona")
plt.xlabel("T")
plt.ylabel("Nt")
plt.grid(True)
plt.show()
```

# Metoda 2

```python
def poiss2(lambd=2, T=10):
    n = np.random.poisson(lambd * T)
    if n != 0:
        U = np.random.uniform(0, T, n)
        S = sorted(U)
    l = len(S)
    N = np.arange(1, l + 1)
    return S, N


S, N = poiss2()


plt.step(S, N)
plt.title("Proces Poissona")
plt.xlabel("T")
plt.ylabel("Nt")
plt.grid(True)
plt.show()
```

```python
n = 1000
Nt = np.zeros(n)
for i in range(n):
    X = poiss1()[1]
    X = len(X)
    Nt[i] = X

Z = np.random.poisson(20, 1000)

sns.ecdfplot(Z, label="Rozkad Poissona")
sns.ecdfplot(Nt, label="Nt")
plt.legend()
plt.show()
```
